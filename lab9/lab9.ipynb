```python
import thinkdsp as tk
import numpy as np
```

## Задание 9.2

```python
wave = tk.TriangleSignal(freq=100).make_wave(duration=0.1, framerate=44100)
wave.plot()
tk.decorate(xlabel='Время (с)')
```

```python
wave_diff = wave.diff()
wave_diff.plot()
tk.decorate(xlabel='Время (с)')
```

```python
wave_differentiate = wave.make_spectrum().differentiate().make_wave()
wave_differentiate.plot()
tk.decorate(xlabel='Время (с)')
```

## Задание 9.3

```python
wave = tk.SquareSignal(freq=100).make_wave(duration=0.1, framerate=44100)
wave.plot()
tk.decorate(xlabel='Время (с)')
```

```python
wave_cum_sum = wave.cumsum()
wave_cum_sum.plot()
tk.decorate(xlabel='Время (с)')
```

```python
spectrum = wave.make_spectrum().integrate()
spectrum.hs[0] = 0
wave_integral = spectrum.make_wave()
wave_integral.plot()
tk.decorate(xlabel='Время (с)')
```

```python
wave_integral.unbias()
wave_integral.normalize()
wave_cum_sum.normalize()
wave_integral.plot()
wave_cum_sum.plot()
```

## Задание 9.5

```python
wave = tk.CubicSignal(freq=100).make_wave(duration=0.05, framerate=44100)
wave.plot()
```

```python
wave_diff1 = wave.diff()
wave_diff1.plot()
```

```python
wave_diff2 = wave_diff1.diff()
wave_diff2.plot()
```

```python
window = np.array([-1.0, 2.0, -1.0])
padded = tk.zero_pad(window, len(wave))
diff_wave = tk.Wave(padded, framerate=wave.framerate)
diff_filter = diff_wave.make_spectrum()
diff_filter.plot(label='2nd diff')
tk.decorate(xlabel='Частота (Гц)', ylabel='Амплитуда')
```

```python
deriv_filter = wave.make_spectrum()
deriv_filter.hs = (np.pi * 2 * 1j * deriv_filter.fs)**2
deriv_filter.plot(label='2nd deriv')
tk.decorate(xlabel='Частота (Гц)', ylabel='Амплитуда')
```

```python
diff_filter.plot(label='2nd diff')
deriv_filter.plot(label='2nd deriv')
tk.decorate(xlabel='Частота (Гц)', ylabel='Амплитуда')
```
